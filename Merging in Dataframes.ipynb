```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
```

```python
df1 = DataFrame({'Name':['Lasya','Madhav','Shravan','Meghana'],
                'Surname':['Bhaskara','Peri','Bhaskara','Kothakonda'],'Age':[26,29,31,28]})
df1
```

```python
df2 = DataFrame({'Name':['Lasya','Madhav','Sudhakar','Kalyani','Lasya'],
                 'Relation':['Self','Husband','Father','Mother','Me'],'Age':[27,29,61,57,26]})
df2
```

```python
pd.merge(df1,df2,on='Name') # default is inner join: intersection of common column's values(keys), their various cominations.
#name and age are the keys here. If you only want to use one key, specify in 'on' parameter.
```

```python
pd.merge(df1,df2,on='Name',suffixes = ['_Main','_Side']) # to change suffixes for common columns. 
```

```python
pd.merge(df1,df2,on='Age',suffixes = ['_Main','_Side'])
```

```python
pd.merge(df1,df2,'left',on='Name') #left join includes all the keys from left dataframe 
# and common keys from right with their combinations
```

```python
pd.merge(df1,df2,'right',on='Name') #right join:all the keys from right dataframe
#and any combinations with common keys from left frame
```

```python
pd.merge(df1,df2,'outer',on='Name') # all the keys from both the dataframes and their combinations
```

# merge on two keys

```python
pd.merge(df1,df2,on=['Age','Name']) # always observe the columns being printed.
```

# merge on indexes

```python
df1
```

```python
df4 = DataFrame(np.arange(9).reshape(3,3),columns=['A','B','C'],index=[26,27,28])
df4
```

```python
pd.merge(df1,df4,left_on='Age',right_index=True) # common keys of key column on left dataframe and index of the right dataframe
```

# for hierarchy index merging

```python
df1
```

```python
df3 = DataFrame(np.arange(16).reshape(4,4),columns=['A','B','C','D'],index=[[26,29,26,28],['Lasya','Madhav','Lasya','high']])
df3
```

```python
pd.merge(df1,df3,left_on=['Age','Name'],right_index=True)
```
