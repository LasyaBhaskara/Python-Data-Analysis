```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
df1= DataFrame({'Gender':['Boy','Girl','Boy','Boy','Girl','Boy','Girl','Girl','Girl','Boy'],'Age':[2,7,2,9,5,5,7,10,10,20],
                            'State':['ON','ON','ON','BC','BC','MB','ON','SK','SK','SK']})
df1.columns.name = 'Info'
df1.index.name = 'Number'
df1
```

```python
df1.duplicated() # the first time it encounters a row, sets false, second time it encounters the same row, sets true
```

```python
df1.drop_duplicates() # drops rows that are duplicated
```

```python
df1
```

```python
df1.drop_duplicates('Age') # drop duplicates based on specific column values even though entire row is not duplicate
```

```python
df1.drop_duplicates(keep='last',ignore_index=True) #after encountering all duplicate rows,
# keeps the last duplicate
#ignore_index=True implies, instead of original indexes from the dataframe, use a new indexing for the resulting dataframe
```
